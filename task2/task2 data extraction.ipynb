```python
import yfinance as yf
import pandas as pd
import re
```

```python
# dataset link : https://finance.yahoo.com/quote/AAPL/history/?period1=1609459200&period2=1767767620
```

```python
with open('raw_data.txt', 'r') as f:
    data = f.read()
```

```python
raw_data = data
```

```python
parts = re.split(r'([A-Z][a-z]{2}\s\d{1,2},\s\d{4})', raw_data)
```

```python
data = []

for i in range(1, len(parts), 2):
    date = parts[i].strip()
    rest = parts[i+1].strip()

    prices = re.findall(r'(\d+\.\d{2})', rest)
    
    current_prices = prices[:5]
    
    temp_rest = rest
    for p in current_prices:
        temp_rest = temp_rest.replace(p, '', 1)
    
    volume = temp_rest.strip()
    
    row = [date] + current_prices + [volume]
    data.append(row)
```

```python
df = pd.DataFrame(data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
df.to_csv('data.csv', index=False)
```
