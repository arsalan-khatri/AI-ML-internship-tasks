```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = sns.load_dataset('iris')
```

```python
# Find dataset Totals rows and Total columns
print("Total Rows :",df.shape[0])
print("Total Columns :",df.shape[1])
print("Total rows and columns :",df.shape)
```

```python
print("Dataset first 5 Rows :\n\n",df.head(),"\n")
print("Dataset first 3 Rows :\n\n",df.head(3))
```

```python
print("Summary Statistics\n\n",df.describe())
```

```python
# Find dataset columns types and null values
print("Dataset Information")
print(df.info())
```

```python
# A Scatter Plot [cite: 18]
# find dependancy 'sepal_length' on 'sepal_width'
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# B. Histograms [cite: 19]
# find values Distribution 
df.hist(figsize=(8,6), bins=20)
plt.suptitle('Feature Distributions')
plt.show()
```

```python
# C. Box Plots [cite: 20]
# Check 'Outliers'
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Outlier Detection with Box Plot')
plt.show()
```
